```python
import sqlite3
import spacy

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# SQLite connection
conn = sqlite3.connect("student_database.db")  # Replace with your database name
cursor = conn.cursor()

# Query map with synonyms/variations
QUERY_MAP = {
    "name  roll number": 'SELECT FullName FROM Students WHERE RegisterNumber = ?',
    "email  roll number": 'SELECT Email FROM Students WHERE RegisterNumber = ?',
    "batch year  roll number": 'SELECT BatchYear FROM Students WHERE RegisterNumber = ?',
    "gender  roll number": 'SELECT Gender FROM Students WHERE RegisterNumber = ?',
    "dob  roll number": 'SELECT DateOfBirth FROM Students WHERE RegisterNumber = ?',
    "phone number  roll number": 'SELECT PhoneNumber FROM Students WHERE RegisterNumber = ?',
    "parent name  roll number": 'SELECT ParentsName FROM Students WHERE RegisterNumber = ?',
    "parent contact  roll number": 'SELECT ParentsContact FROM Students WHERE RegisterNumber = ?',
    "cgpa  roll number": 'SELECT CGPA FROM Students WHERE RegisterNumber = ?',
    "1st sem sgpa  roll number": 'SELECT Sem1_SGPA FROM Students WHERE RegisterNumber = ?',
    "2nd sem sgpa  roll number": 'SELECT Sem2_SGPA FROM Students WHERE RegisterNumber = ?',
    "3rd sem sgpa  roll number": 'SELECT Sem3_SGPA FROM Students WHERE RegisterNumber = ?',
    "4th sem sgpa  roll number": 'SELECT Sem4_SGPA FROM Students WHERE RegisterNumber = ?',
    "blood group  roll number": 'SELECT BloodGroup FROM Students WHERE RegisterNumber = ?',
    "projects  roll number": 'SELECT ListOfProjects FROM Students WHERE RegisterNumber = ?',
    "number of projects  roll number": 'SELECT ProjectsDone FROM Students WHERE RegisterNumber = ?',
    "club membership  roll number": 'SELECT ClubMembership FROM Students WHERE RegisterNumber = ?',
    "club name  roll number": 'SELECT ClubName FROM Students WHERE RegisterNumber = ?',
    "certifications  roll number": 'SELECT Certifications FROM Students WHERE RegisterNumber = ?',
    "linkedin  roll number": 'SELECT LinkedInProfile FROM Students WHERE RegisterNumber = ?',
    "github  roll number": 'SELECT GitHubProfile FROM Students WHERE RegisterNumber = ?',
    "skills  roll number": 'SELECT Skills FROM Students WHERE RegisterNumber = ?',
    "list all students": 'SELECT FullName, RegisterNumber FROM Students',
    "list all emails": 'SELECT FullName, Email FROM Students',
    "list all phone numbers": 'SELECT FullName, PhoneNumber FROM Students',
    "list all blood groups": 'SELECT FullName, BloodGroup FROM Students',
    "list all cgpa": 'SELECT FullName, CGPA FROM Students',
    "roll number by name": 'SELECT RegisterNumber FROM Students WHERE FullName LIKE ?'

}

# Common stop words to ignore
STOP_WORDS = {"what", "is", "the", "by", "of", "?", "please"}

def process_query(user_input):
    """Processes the user's query."""
    # Tokenize input using SpaCy
    doc = nlp(user_input.lower())
    
    # Extract keywords, removing stop words
    tokens = [token.text for token in doc if token.text not in STOP_WORDS]
    print(f"DEBUG: Tokens: {tokens}")
    
    # Match query type
    query_type = None
    for key in QUERY_MAP.keys():
        key_tokens = key.lower().split()
        if all(token in tokens for token in key_tokens):  # Ensure all tokens are present
            query_type = key
            break

    if not query_type:
        return "Query type not recognized. Please check your input."

    # Prepare SQL query
    sql_query = QUERY_MAP[query_type]
    params = []

    # Extract roll number if required
    if "roll number" in query_type:
        roll_number = None
        for token in doc:
            if token.is_digit and len(token.text) >= 6:  # Assuming roll numbers are numeric and have 6+ digits
                roll_number = token.text
                break
        
        if roll_number:
            params.append(roll_number)
        else:
            return "No valid roll number found in your query."
        
    params = []
     # Handle queries with name
    if "name" not in tokens and "roll number" in query_type:
        # Extract name
        name_tokens = [token.text for token in doc if token.is_alpha and token.text not in STOP_WORDS]
        print(f"DEBUG: Name tokens: {name_tokens}")
        full_name = " ".join(name_tokens)  # Join tokens to form the name
        if full_name:
            params.append(f"%{full_name}%")
        else:
            return "No valid name found in your query."

    # Execute SQL query
    try:
        cursor.execute(sql_query, params)
        result = cursor.fetchall()
        return result if result else "No data found."
    except Exception as e:
        return f"Error: {str(e)}"

# Interactive prompt
if __name__ == "__main__":
    print("Student Query System")
    print("Type 'exit' to quit.")
    while True:
        user_query = input("\nEnter your query: ")
        if user_query.lower() == "exit":
            print("Goodbye!")
            break
        response = process_query(user_query)
        print(f"Response: {response}")

# Close the SQLite connection at the end
conn.close()
```
